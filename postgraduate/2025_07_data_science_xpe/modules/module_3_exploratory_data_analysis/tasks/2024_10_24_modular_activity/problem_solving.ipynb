#### Entre os países que ganharam pelo menos 10 medalhas, qual país tem a maior taxa de medalhas per capita?

```python
df_paisesm10_medalhas = df_medalha[df_medalha['Total']>=10]
df_paisesm10_medalhas
```

```python
# Entre os países que ganharam pelo menos 10 medalhas, qual país tem a maior taxa de medalhas per capita?
df_paisesm10_merged = df_paisesm10_medalhas.merge(df_populacao,right_on='Country Code',left_on='country_code',how='left')
df_paisesm10_merged['populacao_por_medalha'] = df_paisesm10_merged['Total']/df_paisesm10_merged['2023']
max_taxa_med_pop= df_paisesm10_merged['populacao_por_medalha'].max()
df_paisesm10_merged.loc[df_paisesm10_merged['populacao_por_medalha']==max_taxa_med_pop]
```

```python
# Step 1: Merge the DataFrames on country code
df_paises_merged = df_paisesm10_medalhas.merge(
    df_populacao, right_on='Country Code', left_on='country_code', how='left'
)

# Step 2: Calculate the population per medal ratio
df_paises_merged['pop_per_medal'] = df_paises_merged['Total'] / df_paises_merged['2023']

# Step 3: Identify the minimum population per medal ratio
min_pop_per_medal = df_paises_merged['pop_per_medal'].min()

# Step 4: Filter to find the country or countries with this minimum ratio
min_pop_per_medal_country = df_paises_merged.loc[df_paises_merged['pop_per_medal'] == min_pop_per_medal]

# Optional: Display the country with the minimum population per medal ratio
print("Country(ies) with minimum population per medal:", min_pop_per_medal_country)

# Step 5: Sort the DataFrame by population per medal ratio for visualization
df_paises_sorted = df_paises_merged.sort_values('pop_per_medal', ascending=True)

# Step 6: Create a bar chart for population per medal by country
plt.figure(figsize=(10, 6))
plt.bar(df_paises_sorted['country_code'], df_paises_sorted['pop_per_medal'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population per Medal')
plt.title('Population per Medal by Country')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust layout for readability
plt.show()
```

#### Q10. Entre os fatores renda per capita, população e renda total (renda total da população), qual fator mais explica o número de medalhas alcançado por um país? 

```python
# Merge entre medalhas e população
df_q10_med_pop = df_medalha.merge(
    df_populacao[['Country Code', '2023']],  # Selecione as colunas corretamente
    right_on='Country Code',
    left_on='country_code',
    how='left'
)

# Merge do resultado anterior com renda per capita
df_q10 = df_q10_med_pop.merge(
    df_renda_percapita[['Country Code', '2023']],  # Selecione as colunas corretamente
    right_on='Country Code',
    left_on='country_code',
    how='left',
    suffixes=('_pop', '_renda')  # Adicionando sufixos para distinguir colunas iguais
)

# Remover a coluna "Country Code"
df_q10 = df_q10.drop(columns=['Country Code_pop','Country Code_renda'])

# Adiciona Renda total
df_q10['total_renda']=df_q10['2023_pop']*df_q10['2023_pop']

# Exibir os resultados
df_q10.head()
```

```python
# Quais variáveis são mais fortemente correlacionados? R: População
corr_q10= df_q10[['2023_pop','2023_renda','total_renda','Total']].corr()

hm = sns.heatmap(
    corr_q10,
    cbar=True,
    annot=True,
    square= True,
    fmt = '.2f',
    annot_kws={'size':5},
    yticklabels=corr_q10.columns,
    xticklabels=corr_q10.columns,
    cmap='Spectral_r'
)
plt.show()
```

#### Q12. Execute o código abaixo e responda qual a faixa de idade em que mais atletas ganharam medalhas.

```python
medallists_df = df_medalistas.copy()
```

```python
medallists_df.head()
```

```python
# Converter a coluna 'birth_date' para datetime
medallists_df['birth_date'] = pd.to_datetime(medallists_df['birth_date'])

# Obter a data atual
today = pd.to_datetime(datetime.today().date())

today
```

```python
# Get the current date
today = pd.to_datetime(datetime.today().date())

# Calculate initial age
medallists_df['age'] = today.year - medallists_df['birth_date'].dt.year

# Check if the birthday has occurred this year and adjust the age
birthday_passed = (today.month > medallists_df['birth_date'].dt.month) | \
                  ((today.month == medallists_df['birth_date'].dt.month) & (today.day >= medallists_df['birth_date'].dt.day))

# Subtract 1 from age if the birthday hasn't occurred yet
medallists_df['age'] -= ~birthday_passed.astype(int)

medallists_df['age']
```

```python
bins = [0, 20, 25, 30, 35, 40, 100]

labels = ['Under 20', '20-24', '25-29', '30-34', '35-39', '40 and over']

medallists_df['age_group'] = pd.cut(medallists_df['age'], bins=bins, labels=labels)

 

age_group_medal_count = medallists_df.groupby(['country', 'age_group'])['medal_type'].count().unstack().fillna(0)

 

top_10_countries_age = age_group_medal_count.sum(axis=1).sort_values(ascending=False).head(10)

age_group_medal_count = age_group_medal_count.loc[top_10_countries_age.index]
```

```python
plt.figure(figsize=(14, 10))

age_group_medal_count.T.plot(kind='bar', stacked=True, figsize=(14, 10), cmap='Set3')

 

plt.xlabel('Idade', fontsize=14)

plt.ylabel('Número de Medalhas', fontsize=14)

plt.title('Desempenho em Medalhas por idade para os 10 paises com mais medalhas', fontsize=18, weight='bold')

plt.legend(title='Pais', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
