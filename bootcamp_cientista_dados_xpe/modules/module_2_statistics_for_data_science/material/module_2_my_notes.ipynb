## BOXPLOT 

lim_superior = Q3+1,5*IQR

Q3 = 75% ABAIXO => SUPERIOR BOXPLOT (Limite superior do corpo)

Q2 = 50% ABAIXO => CENTRO BOXPLOT = mediana -> 'medida do meio, aps ordenao'

Q1 = 25% ABAIXO => INFERIOR BOXPLOT (Limite inferior do corpo)

Q4 = limites inf + sup + outliers

IQR = intervalo inter quartil = Q3-Q1

lim_inferior = Q1-1,5*IQR


Outlier = > limsup ou < lim inf

### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# histograma de vendas de acordo com o preo do caf
plt.hist(df['Preco_Cafe'],bins=6,edgecolor='black')

# Adicionando ttulos e rotulos
plt.title('Distribuio do preo do caf')
plt.xlabel('Preo do caf')
plt.ylabel('Contagem de amostras | Frequncia')

plt.show()
```

```python
# Criar subplots
fig, axs = plt.subplots(3,figsize = (4,6))

# Primeiro subplot
axs[0].hist(df['Vendas_Cafe'],bins = 6, edgecolor = 'black')
axs[0].set_title('Distribuio de vendas de Caf')
axs[0].set_xlabel('Vendas de Caf')
axs[0].set_ylabel('Frequncia')

# Segundo subplot
axs[1].hist(df['Preco_Cafe'],bins = 6, edgecolor = 'black')
axs[1].set_title('Distribuio de preos do Caf')
axs[1].set_xlabel('Preo do Caf')
axs[1].set_ylabel('Frequncia')

# Terceiro subplot
axs[2].scatter(df['Vendas_Cafe'],df['Preco_Cafe'])
axs[2].set_title('Vendas do Caf x Preo do Caf')
axs[2].set_xlabel('Vendas do Caf')
axs[2].set_ylabel('Preo do Caf')

plt.show()
```

```python
#--------------
#Separando dataframe com e sem promocao
dadosNao = df [df['Promocao']=="Nao"]

dadosSim = df [df['Promocao'] =="Sim"]

print(dadosNao)

print(dadosSim)


plt.title('Relao entre Preo do Caf e Vendas de Caf')
plt.xlabel('Preo do Caf')
plt.ylabel('Vendas de Caf')

plt.scatter(dadosSim['Preco_Cafe'], dadosSim['Vendas_Cafe'], color = 'red', label = 'Com promoo')
plt.scatter(dadosNao['Preco_Cafe'], dadosNao['Vendas_Cafe'], color = 'blue', label='Sem promoo')

plt.legend()

# Mostrar o grfico
plt.show()
```

```python

#--------------
# Gerar boxplot das vendas de caf
plt.boxplot(df['Vendas_Cafe'])

# Adicionar ttulos e rtulos
plt.title('Boxplot das Vendas de Caf')
plt.ylabel('Vendas de Caf')

# Configurar os rtulos do eixo x para mostrar que representam as Vendas de Caf
plt.xticks([1], ['Vendas de Caf'])
plt.show()


#Mesma coisa para Preo do Caf
plt.boxplot(df['Preco_Cafe'])

# Adicionar ttulos e rtulos
plt.title('Boxplot dos Preos de Caf')
plt.ylabel('Preos de Caf')

# Configurar os rtulos do eixo x para mostrar que representam as Vendas de Caf
plt.xticks([1], ['Preos de Caf'])
plt.show()
```

```python

#--------------

#Plota boxplot com dados de promocao No e sim
plt.boxplot([dadosNao['Vendas_Cafe'], dadosSim['Vendas_Cafe']])
plt.title('Boxplot das Vendas de Caf por Promoo')
plt.xticks([1,2], ['No','Sim'])
plt.ylabel('Vendas de Caf')
plt.show()
```
